```python
import pandas as pd
import seaborn as sns
```

```python
# Bivariate And Multivariate Analysis:- Bivariate is used to do analysis of 2 vriable.
# multivariate is used to do analysis of more than two columns
```

```python
tips = sns.load_dataset('tips')
```

```python
titanic = pd.read_csv('Titanic.csv')
```

```python
flights = sns.load_dataset('flights')
```

```python
iris = sns.load_dataset('iris')
```

# 1. Scatterplot (Numerical-Numerical)

```python
sns.scatterplot(x=tips['total_bill'],y=tips['tip'],hue=tips['sex'])
```

# 2. BarPlot (Numerical-Categorical)

```python
# Pclass with Age
```

```python
sns.barplot(x=titanic['Pclass'],y=titanic['Age'],hue=titanic['Survived'])
```

```python
# Pclass With Fare
```

```python
sns.barplot(x=titanic['Pclass'],y=titanic['Fare'],hue=titanic['Sex'])
```

# 3. Box Plot (Numerical-Categorical)

```python
sns.boxplot(x=titanic['Sex'],y=titanic['Age'])
```

```python
# 4. DistPlot (Numerical-Categorical)
```

```python
sns.distplot(titanic[titanic['Survived']==0]['Age'],hist=False) # died
sns.distplot(titanic[titanic['Survived']==1]['Age'],hist=False) # survived
```

# 5. Heat Map (Categorical - Categorical)

```python
pd.crosstab(titanic['Pclass'],titanic['Survived']) # crosstab is used to show relation between category vs category
```

```python
sns.heatmap(pd.crosstab(titanic['Pclass'],titanic['Survived']))
```

```python
flights
```

```python
flights.pivot_table(values='passengers',index='month',columns='year')  # pivot table is used to show relation between category and numerical data
```

```python
sns.heatmap(flights.pivot_table(values='passengers',index='month',columns='year'))
```

# 6. Pair Plot 

```python
sns.pairplot(iris,hue='species')
```

# 7. Line Plot (Numerical-Numerical)

```python
new = flights.groupby('year').sum().reset_index()
```

```python
sns.lineplot(new['year'],new['passengers'])
```
